# Text Input

https://platform.openai.com/docs/models


```python
from langchain.agents import create_agent

agent = create_agent(
    model="gpt-5-nano",
    system_prompt="You are a DJ for almost 2 decades and played in every part of the world in famous venues, festivals and genres. If user request you about any type of information on your field, you must answer from your experience."
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content=[
    {"type":"text", "text": "How can I start learning DJ ?"}
])

response = agent.invoke(
    {"messages":[question]}
)

print(response["messages"][-1].content)
```

# Image Input

Here we need to first upload Image, so need to write an upload function.

Then Image will be converted to base64

```python
from ipywidgets import FileUpload
from IPython.display import display

uploader = FileUpload(accept=".png", multiple=False)
display(uploader)
```

```python
print(uploader.value)
```

```python
import base64

# Get the first (and only) uploaded file dict
uploaded_file = uploader.value[0]

# This is a memory view
content_mv = uploaded_file["content"]

# Content memoryview -> bytes
img_bytes = bytes(content_mv)

# Now base64 encode
img_b64 = base64.b64encode(img_bytes).decode("utf-8")
```

```python
multimodal_question = HumanMessage(content=[
    {"type":"text", "text":"Tell me about this guy, should he be able to DJ on stage based on its appearance and attire"},
    {"type": "image", "base64": img_b64, "mime_type": "image/png"}
])

response = agent.invoke(
    {"messages":[multimodal_question]}
)

print(response["messages"][-1].content)
```

# Audio Input

```python
import sounddevice as sd
from scipy.io.wavfile import write
import base64
import io
import time
from tqdm import tqdm

# recording settings
duration = 5 # seconds
sample_rate = 44100

print("Recording...")
audio = sd.rec(int(duration*sample_rate), samplerate=sample_rate, channels=1)

# Progress bar for the duration
for _ in tqdm(range(duration*10)): # update 10x per second
    time.sleep(0.1)
sd.wait()
print("Done.")


# Write WAV to an in-memory buffer
buf = io.BytesIO()
write(buf, sample_rate, audio)
wav_bytes = buf.getvalue()

aud_b64 = base64.b64encode(wav_bytes).decode("utf-8")
```

```python
agent = create_agent(
    model="gpt-4o-audio-preview"
)

multimodal_question = HumanMessage(content=[
    {"type":"text", "text":"Tell me about this audio file"},
    {"type":"audio", "base64":aud_b64, "mime_type":"audio/wav"}
])

response = agent.invoke(
    {"messages":[multimodal_question]}
)

print(response['messages'][-1].content)
```
