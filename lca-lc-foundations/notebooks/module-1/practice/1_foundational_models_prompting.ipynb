# Streaming Output

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about Solis city, the capital of the Saturn")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content: # Check if there's actual content
        print(token.content, end="", flush=True) # Print token 
```

# Role based Conversation (specific niche)

```python
system_prompt = "You are a science fiction writer, create a capital city at the users request."

question = HumanMessage(content="What's the capital of Saturn?")

scifi_agent = create_agent(
    model="gpt-5-nano",
    system_prompt=system_prompt
)

response = scifi_agent.invoke(
    {"messages": [question]}
)

print(response["messages"][1].content)
```

If we check above we have too long response, to handle the response we can provide FEW-SHOT examples. 

# Few-shot examples

```python
system_prompt = """

You are a science fiction writer, create a capital city at the users request.

User: What is the capital of Moon?
Scifi Writer: Luna City

User: What is the capital of Mars?
Scifi Writer: Olympus City

"""

scifi_agent = create_agent(
    model="gpt-5-nano",
    system_prompt=system_prompt
)

response = scifi_agent.invoke({
    "messages": [HumanMessage(content="What's the capital of Saturn?")]
})

print(response["messages"][1].content)
```

# Structured Prompts

```python
system_prompt = """

You are a science fiction writer, create a space capital city at the users request.

Please keep to the below structure.

Name: The name of the capital city.

Location: Where it is based.

Vibe: 2-3 words to describe its vibe

Economy: Main industries.

"""

scifi_agent = create_agent(
    model="gpt-5-nano",
    system_prompt=system_prompt,
)

response = scifi_agent.invoke({
    "messages": [question]
})

print(response["messages"][1].content)
```

# Structured Output

```python
from langchain.agents import create_agent
from langchain.messages import HumanMessage
from pydantic import BaseModel

class CapitalInfo(BaseModel):
    name: str
    location: str
    vibe: str
    economy: str


agent = create_agent(
    model="gpt-5-nano",
    system_prompt=system_prompt,
    response_format=CapitalInfo
)

question = HumanMessage(content="What is the capital of Saturn?")

response = agent.invoke({
    "messages": [question]
})


response["structured_response"]
```

```python
response["structured_response"].name
```

```python
response["structured_response"].economy
```
