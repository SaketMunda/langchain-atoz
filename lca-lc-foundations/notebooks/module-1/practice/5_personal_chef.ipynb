# Goal

In this notebook, we will be going to test all the modules we have learned so far in Langchain. 

To demonstrate the skill we will be building a personal chef agents, which takes inputs as the ingredients left over in the kitchen, do the web search to find what we can cook based on the ingredients and then propose a dish.

```python
# import the env
from dotenv import load_dotenv
load_dotenv()
```

Let's start by importing packages

```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain.messages import HumanMessage
from tavily import TavilyClient
from langgraph.checkpoint.memory import InMemorySaver 
from pprint import pprint
```

```python
# define the role
system_prompt = "You are a chef of a Michelin Star Restraunt and your task is to innovate dishes based on the ingredients and tools provided."

# create the TavilyClient instance
tavily_client = TavilyClient()

# declare the tool for web search
@tool
def web_search(query: str) -> str:
    """
        This tool is responsible to do the web search based on the query requested as input and return the result.
    """
    return tavily_client.search(query=query)

# Create a short term memory config so that it will remember what we are doing in the session
config = {"configurable": {"thread_id": "1"}}

# initialise the agent
agent = create_agent(
    model="claude-sonnet-4-5",
    system_prompt=system_prompt,
    tools=[web_search],
    checkpointer=InMemorySaver()
)
```

```python
# Let's invoke the agent
question = HumanMessage(content="Hey, I've left raw boneless chicken 250gm, brocolli, onion, carrot, sweetpotato in my fridge.")

response = agent.invoke(
    {"messages": [question]},
    config=config
)


pprint(response["messages"][-1].content)
```
