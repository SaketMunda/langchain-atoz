# No Memory

```python
from langchain.agents import create_agent

agent = create_agent(
    model="gpt-5-nano"
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello my name is 5aketx and I'm a DJ.")

response = agent.invoke({
    "messages": [question]
})


print(response)
```

```python
from pprint import pprint

pprint(response["messages"])
```

```python
question = HumanMessage(content="Who am I?")

response = agent.invoke(
    {"messages":[question]}
)

pprint(response["messages"])
```

# Memory

```python
from langgraph.checkpoint.memory import InMemorySaver

agent = create_agent(
    model="gpt-5-nano",
    checkpointer=InMemorySaver(),
)
```

```python
from langchain.messages import HumanMessage

question = HumanMessage(content="Hello, My name is 5aketx and I'm pursuing DJ.")
config = {"configurable": {"thread_id": "1"}}

response = agent.invoke(
    {"messages":[question]},
    config,
)
```

```python
pprint(response)
```

```python
question = HumanMessage(content="Who am I?")

response = agent.invoke(
    {"messages": [question]},
    config
)

pprint(response)
```

```python
question = HumanMessage(content="What I do?")

response = agent.invoke(
    {"messages": [question]},
    config
)

pprint(response)
```
